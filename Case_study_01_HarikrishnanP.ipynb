```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Read the dataset to the python environment.

```python
df=pd.read_csv('cars_data.csv')
df
```

## 2. Check for the null values present in the dataset.


```python
df.isnull().sum()
```

## 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
ax=sns.countplot(x='Buyer Gender',data=df)
```

## 4. Find the top 5 cars based on their sales price.

```python
df1=df[['Make','Sale Price']]
df_top5 = df1.sort_values(by='Sale Price',ascending=False).head()
print("Top 5 cars based on sales price are :")
print(df_top5.to_string(index=False))

df_top5.plot(x='Make',y='Sale Price',kind='bar',color='green',legend=False)
plt.xlabel('Car Make')
plt.ylabel('Sale Price')
plt.title('Car Make vs Sale Price')
```

## 5. Find the least 5 cars based on their Resell price.


```python
df2=df[['Make','Resell Price']]
df_least5 = df2.sort_values(by='Resell Price',ascending=True).head()
print("Least 5 cars based on Resell price are :")
print(df_least5.to_string(index=False))

df_least5.plot(x='Make',y='Resell Price',kind='bar',color='green',legend=False)
plt.xlabel('Car Make')
plt.ylabel('Resell Price')
plt.title('Car Make vs Resell Price')
```
